```python
import pandas as pd
from ast import literal_eval
import matplotlib.pyplot as plt
import seaborn as sns
```

# Películas


```python
peliculas = pd.read_csv('./starz_peliculas.csv')
```

```python
peliculas.head(n=10)
```

```python
top_ten = peliculas.sort_values(by=['Duracion(min)'],ascending=False).iloc[:10]
plt.figure(figsize=(12,12))
ax = sns.barplot(x="Titulo", y="Duracion(min)", data=top_ten)
plt.xticks(rotation=70)
plt.tight_layout()
ax.set_title('Top Ten de Películas según su duración')
```

```python
top_ten = peliculas.sort_values(by=['Duracion(min)']).iloc[:10]
plt.figure(figsize=(12,12))
ax = sns.barplot(x="Titulo", y="Duracion(min)", data=top_ten)
plt.xticks(rotation=70)
plt.tight_layout()
ax.set_title('Top Ten de Películas según menor duración')
```

```python
plt.figure(figsize=(10,10))
ax = sns.barplot(x=peliculas.Calificacion.value_counts().index, y=peliculas.Calificacion.value_counts())
plt.xticks(rotation=70)
plt.tight_layout()
ax.set_title('Cantidad de Peliculas según su Calificación')
```

```python
#histograma de años de lanzamiento de las películas
peliculas.hist(column='Año')
```

# Series

```python
#utilizamos la función literal_eval para serializar el diccionario de info_episodios y que no se guarde como str
series = pd.read_csv('./starz_series.csv', converters={'info_episodios': literal_eval})
```

```python
series.head(n=10)
```

```python
top_ten = series.sort_values(by=['Episodios'],ascending=False).iloc[:10]
plt.figure(figsize=(12,12))
ax = sns.barplot(x="Titulo", y="Episodios", data=top_ten)
plt.xticks(rotation=70)
plt.tight_layout()
ax.set_title('Top Ten de Series según su cantidad de Episodios')
```

```python
top_ten = series.sort_values(by=['Temporadas'],ascending=False).iloc[:10]
plt.figure(figsize=(12,12))
ax = sns.barplot(x="Titulo", y="Temporadas", data=top_ten)
plt.xticks(rotation=70)
plt.tight_layout()
ax.set_title('Top Ten de Series según su cantidad de Temporadas')
```

### Creamos un nuevo df con la info de los episodios

```python
lista_episodios = []
for index, row in series.iterrows():
    lista_dicc = row['info_episodios']
    
    for episodio in lista_dicc:
        episodio['Serie'] = row['Titulo']
        lista_episodios.append(episodio)
        
```

```python
df_ep = pd.DataFrame(lista_episodios)
```

```python
df_ep.head()
```

#### Verificamos que no se hayan scrapeado tráilers

```python
titulos = list(df_ep['Titulo'])
trailers = 0
for titulo in titulos:
    if 'tráiler' in titulo.lower():
        print(titulo)
        trailers +=1

print(f'Hay {trailers} trailer')
```

```python
#histograma de duración de los episodios
df_ep.hist(column='Duracion(min)')
```

```python
duracion_promedio = df_ep.groupby(['Serie']).mean().sort_values(by=['Duracion(min)']).iloc[:10]
plt.figure(figsize=(12,12))
ax = sns.barplot(x=duracion_promedio.index, y="Duracion(min)", data=duracion_promedio)
plt.xticks(rotation=70)
plt.tight_layout()
ax.set_title('Top Ten de Series según la duración promedio de sus episodios')
```
